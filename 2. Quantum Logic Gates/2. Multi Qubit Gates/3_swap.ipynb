# Controlled SWAP (cswap) Gate
- Conditionally swaps the two target qubits if the controll qubit is ket 1
- SWAP|000> = |000>
- SWAP|001> = |001>
- SWAP|010> = |010>
- SWAP|011> = |011>
- SWAP|100> = |100>
- SWAP|101> = |110>
- SWAP|110> = |101>
- SWAP|111> = |111>

```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram
from math import pi
%matplotlib inline
```

```python
ckt = QuantumCircuit(3)
ckt.x(0)
ckt.ry(pi/2, 1)
ckt.rx(-pi/3, 2)
ckt.draw(output='mpl')
```

```python
# Before swapping
simulator = Aer.get_backend('statevector_simulator')
result = execute(ckt, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```

```python
# After swapping
ckt.cswap(0, 1, 2)
ckt.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(ckt, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```
