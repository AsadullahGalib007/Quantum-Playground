# SWAP Gate
- Will swap the quantum state of two qubits if the two qubits do not match
- SWAP|00> = |00>
- SWAP|01> = |10>
- SWAP|10> = |01>
- SWAP|11> = |11>


```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram
from math import pi
%matplotlib inline
```

```python
ckt = QuantumCircuit(2)
ckt.x(0)
# circuit.h(1)
ckt.rx(-pi/4, 1)
ckt.draw(output='mpl')
```

```python
# Before swapping
simulator = Aer.get_backend('statevector_simulator')
result = execute(ckt, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```

```python
# After swapping
ckt.swap(0,1)
ckt.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(ckt, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```
