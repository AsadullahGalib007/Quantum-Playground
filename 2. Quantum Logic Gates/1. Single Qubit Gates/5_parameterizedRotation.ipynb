```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
from math import pi
%matplotlib inline
```

# Paraeterized Rotation
- Rx, Ry, Rz - Rotate the Bloch Sphere to a specified angle around the X, Y, and Z axis respectively.
- Applying Pi radians to the Rx, Ry, and Rz gates work the same as Pauli-X, Y and Z gates.

```python
ckt = QuantumCircuit(4)
ckt.rx(pi/2, 1)
ckt.ry(-pi/3, 2)
ckt.h([0,3])
ckt.rz(pi/3, 3)
ckt.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(ckt, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```
