# Phase shift 
- S Gate: This single qubit superposition gate applies a phase shift of pi/2 radians, rotating our qubit one quarter turn around the Z-axis of the Bloch sphere.
- S Dagger: Shifts the phase of 3pi/2 radians or 270 degree (-90 degree or - pi/2 rads) around the Z-axis.
- T Gate: Shifts the phase of pi/4 radians or 45 degree around the Z-axis.
- T Dagger: Shifts the phase of -45 degree or -pi/4 radians around the Z-axis.
- S(dot)S dagger = I and T(dot)T dagger = I  

```python
ckt = QuantumCircuit(5)
ckt.h([0, 1, 2, 3, 4])
ckt.s(1)
ckt.sdg(2)
ckt.t(3)
ckt.tdg(4)
ckt.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(ckt, backend=simulator).result()
statevector = result.get_statevector()
```
